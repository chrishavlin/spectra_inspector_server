```python
import matplotlib.pyplot as plt

from spectra_inspector_server._file_tree_handling import EDAXPathHandler
from spectra_inspector_server.processor.operations import OperationEDAXStateHandler

ph = EDAXPathHandler()
ops = OperationEDAXStateHandler(ph)
```

## single channel image

```python
im = ops.get_image("C-12", 300)
plt.imshow(im)
plt.title("single channel image")
plt.show()
```

```python
im = ops.get_image(
    "C-12",
    300,
    index0_range=(300, 600),
    index1_range=(400, 800),
)
plt.imshow(im)
plt.title("single channel image: subset")
plt.show()
```

## multi-channel intensity image

```python
im = ops.get_multi_channel_intensity_image("C-12", channel_range=(200, 400))
plt.imshow(im)
plt.title("multi channel image")
plt.show()
```

```python
im = ops.get_multi_channel_intensity_image(
    "C-12",
    channel_range=(200, 400),
    index0_range=(300, 600),
    index1_range=(400, 800),
)
plt.imshow(im)
plt.title("multi channel image: subset")
plt.show()
```
